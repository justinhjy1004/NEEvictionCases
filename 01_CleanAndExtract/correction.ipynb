This is primarily used to correct parts of the AllEvictionCases.csv data

```python
import pandas as pd
import cleaner
import numpy as np
```

```python
df = pd.read_csv("../Data/AllEvictionCases.csv")
```

```python
df["WRIT_SERVED"], df["WRIT_SERVED_DATE"] = zip(*df.apply(lambda x: cleaner.writ_served(x.WRIT, x.ACTIONS), axis=1))
```

```python
df["HAS_LIMITED_REPRESENTATION"], df["LIMITED_REPRESENTATION_STATUS"], df["LIMITED_REP_ATTORNEY"] = zip(*df.apply(lambda x: cleaner.limited_rep_attorney(x.PARTIES, x.ACTIONS), axis=1))
```

```python
df.groupby(["YEAR", "COURT"]).agg(np.mean)
```

```python
df.to_csv("../Data/AllEvictionCases.csv", index=None)
```
